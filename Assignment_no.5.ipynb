# Q.no.1. 

**Vertical bar for 5 selected years**

```python
import numpy as np
import matplotlib.pyplot as mp

years = ['1003', '1005', '1007', '1009', '1011']
sachin_runs = [1141, 1234, 1005, 915, 1356]
kohli_runs =  [0,    0,    325, 789, 1380]
sehwag_runs = [1250, 990, 1103, 1025, 987]

x = np.arange(len(years))
width = 0.25

mp.figure(figsize=(8, 5))
mp.bar(x - width, sachin_runs, width, label='Sachin')
mp.bar(x, sehwag_runs, width, label='Sehwag')
mp.bar(x + width, kohli_runs, width, label='Kohli')

mp.xlabel('Year')
mp.ylabel('Runs')
mp.title('Runs over Selected Years')
mp.xticks(x, years)
mp.legend()
mp.grid(axis='y', linestyle='--', alpha=0.5)
mp.tight_layout()
mp.Figure (figsize = (25,21))
mp.show()
```

 **Horizontal Bar for total runs scored in their debut 5 years.**

```python
sachin_5yr = [215, 900, 1100, 1300, 1100]  # 1989–1993
sehwag_5yr = [250, 850, 1000, 950, 980]    # 1999–1003
kohli_5yr =  [0, 400, 1010, 1250, 1100]    # 1008–1012

total_runs = [
    sum(sachin_5yr),
    sum(sehwag_5yr),
    sum(kohli_5yr)
]

players = ['Sachin', 'Sehwag', 'Kohli']
colors = ['blue', 'green', 'Red']

mp.figure(figsize=(7, 4))
mp.barh(players, total_runs, color=colors)
mp.xlabel('Total Runs')
mp.title('Total Runs in First 5 Years of Career')
mp.grid(axis='x', linestyle='--', alpha=0.5)
mp.tight_layout()
mp.show()
```

# Q.no.2.

**Pie Chart for Marketing Shares of OS.**

```python
import matplotlib.pyplot as pie

labels = ['Android', 'iOS', 'Others']
shares = [65, 25, 10]
colors = ['#79DD3F', '#e4ecf1', '#96CAEB'] 

pie.figure(figsize=(5, 5))
pie.pie(
    shares,
    labels=labels,
    colors=colors,
    startangle=90,
    autopct='%1.1f%%'  # Shows percentages
)
pie.title('Mobile OS Market Share (Pie Chart)')
pie.axis('equal') 
pie.show()
```

**Horizontal Bar for Marketing Shares of OS.**

```python
pie.figure(figsize=(6, 3.5))
bars = pie.barh(labels, shares, color=colors)

for bar in bars:
    width = bar.get_width()
    pie.text(width + 1, bar.get_y() + bar.get_height() / 2,
             f'{width}%', va='center', fontsize=10)

pie.title('Mobile OS Market Share (Horizontal Bar Chart)')
pie.xlabel('Market Share (%)')
pie.grid(axis='x', linestyle='--', alpha=0.5)
pie.tight_layout()
pie.show()
```

# Which one is easier to understand ?

**Conclusion:**

*For 2–3 categories, the pie chart is visually appealing and fine.

*For precise comparison, especially with small values (like "Others"),
 the horizontal bar chart is clearly more effective.

# Q.no.6.

**Multiple Lines in One Plot: Plot three mathematical functions (sin, cos, exp) on the same axes, each with different colors and line styles. Add a legend.**

```python
import matplotlib.pyplot as jig
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x/3) #avoid steep growth

jig.figure(figsize=(8, 6))
jig.plot(x, y_sin, color='blue', linestyle='-', label='sin(x)')
jig.plot(x, y_cos, color='green', linestyle='-.', label='cos(x)')
jig.plot(x, y_exp, color='red', linestyle=':', label='exp(x/3)')

jig.xlabel("x")
jig.ylabel("y")
jig.title("Multiple Functions: sin(x), cos(x), exp(x/3)")
jig.legend()
jig.grid(True)

jig.show()
```

# Q.no.8.

**Subplots Grid: Create a 2x2 grid of subplots: line plot, bar chart, scatter plot, and histogram, each in a different subplot. Add an overall (supertitle) for the figure.**

```python
import matplotlib.pyplot as jig
import numpy as np

x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

Animes = ['One Piece', 'Naruto', 'Bleach', 'Pokemon', 'DBZ']
networth = [21, 15, 7.5, 92, 28]

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

hist_data = [96, 82, 75, 91, 67, 88, 59, 3, 84, 62, 
        90, 77, 68, 94, 56, 7, 7, 91, 70, 85, 
        92, 66, 74, 0, 97, 94, 60, 76, 83, 
        93, 78, 71, 58, 86, 72, 3, 69, 65, 98, 
        57, 99, 55, 81, 53, 52, 54, 1, 50, 100]

fig, axs = jig.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x_line, y_line, color='#398DF5')
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(Animes, networth, color='orange')
axs[0, 1].set_title("Animes Networth in Billion $ :(Bar Chart)")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='green',s=50)
axs[1, 0].set_title("Scatter Plot")

# Histogram
axs[1, 1].hist(hist_data, bins=5, color='purple', edgecolor='black')
axs[1, 1].set_title("Histogram")

# Supertitle
fig.suptitle("2x2 Grid of Various Plots", fontsize=16)

# Adjust layout to prevent overlap
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plot
jig.show()
```

# Q.no.9

**Stack Plot: Plot the cumulative sales of 3 products over 12 months as a stack plot using made-up data.**

```python
import matplotlib.pyplot as jig
import numpy as np

months = np.arange(1, 13)

product_A = [10, 15, 10, 18, 22, 25, 28, 30, 35, 38, 40, 45]
product_B = [5, 8, 12, 10, 14, 16, 18, 10, 22, 24, 26, 28]
product_C = [2, 4, 6, 7, 8, 9, 10, 12, 13, 14, 16, 18]

jig.figure(figsize=(9, 5))
jig.stackplot(months, product_A, product_B, product_C,
              labels=['Apple', 'Samsung', 'others'],
              colors=['#e4ecf1', '#1F7DF0', '#F0EDAB'])

jig.xlabel("Month")
jig.ylabel("Sales")
jig.title("Cumulative Sales of 3 Products Over 12 Months")
jig.legend(loc='upper left')
jig.grid(True, linestyle='--', alpha=0.7)

jig.show()
```

**Final Sales Ranking (from highest to lowest):**

1. Apple

2. Samsung

3. Others

# Q.no.11.

**Pie Chart with Explode: Create a pie chart showing the market share of four companies. Explode the largest segment and display percentage values on the chart.**

```python
import matplotlib.pyplot as jig

companies = ['Apple', 'Samsung', 'Xiaomi', 'Others']
market_share = [35, 30, 10, 15]
colours=['#e4ecf1','#1F7DF0','#F8B16A','#96CAEB']

explode = [0.1, 0, 0, 0] # 'explode' Apple

jig.figure(figsize=(8, 6))
jig.pie(market_share, labels=companies,
        explode=explode,
        colors=colours,
        shadow=True,
        autopct='%1.1f%%',
        startangle=90)

jig.title('Smartphone Market Share')

jig.axis('equal')

jig.show()
```

# Q.no.12.

**Axis and Legend Customization: For a random line plot, reverse the x-axis, move the legend outside the plot, and increase legend font size.**

```python
import matplotlib.pyplot as jig
import numpy as np

x = np.arange(0, 10, 1) #0-9
y1 = np.random.randint(10, 50, size=10)
y2 = np.random.randint(10, 60, size=10)

# Plot lines
jig.plot(x, y1, label='Jinwoo', color='#045BC6')
jig.plot(x, y2, label='Beru', color='#BB3305')

# Reverse x-axis
jig.gca().invert_xaxis() # gca=gets current axis🗣️

jig.title("Random Line Plot with Custom Axis and Legend")
jig.xlabel("X-axis")
jig.ylabel("Y-axis")

# Move legend outside the plot (right side) & increase font size
jig.legend(loc='center left', bbox_to_anchor=(1, 0.9), fontsize='large')

jig.tight_layout()

jig.show()
```
